*   기간: 2023-11-10 ~ 2023-11-16 (총 7일)
*   리더: 정혜영
*   정의: 마포구 내 공공 흡연부스 설치를 위한 데이터 베이스 설계 
*   목적: 마포구 내 유동인구을 고려하여 설치 위치 후보군을 두고 유초등생 등하교길을 배재한 공공 흡연부스 설치
